```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sklearn
from sklearn.impute import SimpleImputer
warnings.simplefilter("ignore")
```

```python
df = pd.read_excel("datasettt.xlsx")
```

```python
df.describe()
```

```python
start_col = 0
end_col = 2

x = df.iloc[:, 0:3]
y = df.iloc[:, 3]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train_, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer # Ini sudah diimpor di atas, tapi tidak masalah jika diimpor lagi
imputer = SimpleImputer(strategy='mean') # Objek imputer baru dibuat di sini
model = RandomForestClassifier(n_estimators=100, random_state=42)
x_train = imputer.fit_transform(x_resampled)
model.fit(x_train, y_resampled)
```

```python
y_predict = model.predict(x_test)
akurasi = sklearn.metrics.accuracy_score(y_test, y_predict)
```

```python
results = pd.DataFrame({
    'Kunjungan PerMinggu': x_test['Kunjungan PerMinggu'],
    'Waktu Rata-Rata Situs(menit)': x_test['Waktu Rata-Rata Situs(menit)'],
    'Predicted': y_predict,
    'Actual': y_test
})
print(results)

results.to_excel('hasil_PREDIC_STROKE.xlsx', index=False)
```
